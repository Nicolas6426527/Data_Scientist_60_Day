### 2. 請將 Dcard API 取得所有的看板資訊轉換成 DataFrame，並且依照熱門程度排序後存成一個 csv 的檔案。


```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums')
response = r.text

import json
data = json.loads(response)

print(data)
```

```python
'''
Your Code
'''
df = pd.DataFrame(data)
df.sort_values(by="subscriptionCount",ascending=False,inplace=True)#用subscriptionCount當排序，ascending＝False >> 降序（大到小）
```

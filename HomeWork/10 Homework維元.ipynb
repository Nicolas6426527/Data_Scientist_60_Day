df.loc[ '2013-01-01', 'A']

````python
# 記得先 Import 正確的套件

'''
Your Code
'''
import pandas as pd
import numpy as np
````

````python
df = pd.DataFrame(np.random.randn(6,3),columns=['col1','col2','col3'])
df
````

````python
'''
Your Code
'''
print(df.iloc[:3,:])
print(df.iloc[[1,3,5],:])
print(df.iloc[:,1:3])
print(df.iloc[:,0::2])
````

### 3. 請根據提供的資料，選擇出下列的要求：

```
# - 1. filtered by first column > 20?
# - 2. filtered by first column + second column > 50
# - 3. filtered by first column < 30 or second column > 30
# - 4. filtered by total sum of row > 100
```

````python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
````

````python
'''
Your Code
'''
print(df.loc[df[0]>20,0])
print(df.loc[(df[0]+df[1])>50,[0,1]])
print(df.loc[(df[0]<30)|(df[1]>30),[0,1]])
print(df[df.sum(axis=1)>100])
````
